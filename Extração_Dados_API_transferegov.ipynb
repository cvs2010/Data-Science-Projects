```python
import requests
import zipfile
import os

def download_and_extract_zip(url, destination_folder="."):
    """
    Downloads a ZIP file from a URL and extracts it.

    Args:
        url: The URL of the ZIP file.
        destination_folder: The folder to save and extract the file.
    """
    # Create the folder if it doesn't exist
    os.makedirs(destination_folder, exist_ok=True)

    # Extract the filename from the URL
    filename = os.path.basename(url)
    filepath = os.path.join(destination_folder, filename)

    # Download the file
    response = requests.get(url, stream=True)
    response.raise_for_status()  # Check if the request was successful

    with open(filepath, 'wb') as file:
        for chunk in response.iter_content(chunk_size=8192):
            file.write(chunk)

    # Extract the ZIP file
    with zipfile.ZipFile(filepath, 'r') as zip_ref:
        zip_ref.extractall(destination_folder)

    print(f"File downloaded and extracted to: {destination_folder}")

# URL of the ZIP file
zip_url = "http://repositorio.dados.gov.br/seges/detru/siconv.zip"

# Call the function to download and extract the file
download_and_extract_zip(zip_url)
```

```python
import os
import zipfile

# Caminho do diret√≥rio onde os arquivos CSV est√£o localizados
csv_directory = '/content/mg'

# Nome do arquivo ZIP de sa√≠da
zip_filename = 'transferegov_data_transferencias_especiais_estado_mg_2024.zip'

# Criar um arquivo ZIP e adicionar todos os arquivos CSV do diret√≥rio, excluindo a pasta /content/sample_data
with zipfile.ZipFile(zip_filename, 'w') as zipf:
    for root, dirs, files in os.walk(csv_directory):
        # Excluir a pasta /content/sample_data
        if 'sample_data' in dirs:
            dirs.remove('sample_data')
        for file in files:
            if file.endswith('.csv'):
                zipf.write(os.path.join(root, file), arcname=file)

print(f"Todos os arquivos CSV de {csv_directory} foram salvos em {zip_filename}")
```
